## Training and Testing Data

```python
customers[1:3]
```

```python
X = customers[['Avg. Session Length','Time on App','Time on Website','Length of Membership']]
y = customers['Yearly Amount Spent']
```

** test_size=0.3**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

## Training the Model

```python
from sklearn.linear_model import LinearRegression
```

```python
correlation_dataframe = customers[['Avg. Session Length','Time on App','Time on Website','Length of Membership', 'Yearly Amount Spent']]
```

```python
correlation_dataframe.corr()
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

**Coefficients of the model**

```python
print(lm.coef_)
```

## Predicting Test Data

```python
predictions = lm.predict(X_test)
```

```python
sns.scatterplot(x=y_test,y=predictions)
#plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

## Evaluating the Model

Our model performance by calculating the residual sum of squares and the explained variance score (R^2).

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test,predictions))
print('MSE:', metrics.mean_squared_error(y_test,predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```

## Residuals

```python
#customers[['Yearly Amount Spent']].plot(kind='hist')
sns.distplot((y_test-predictions),bins=50);
# error = y_test - predictions
```

```python
fitted_value= lm.predict(X_train)
residuals= y_train-fitted_value

plt.scatter(fitted_value, residuals)
plt.plot([800,0], [0,0],  'r')
plt.xlabel("fitted value")
plt.ylabel("residuals")

plt.xlim(0,800)
plt.show()
```
